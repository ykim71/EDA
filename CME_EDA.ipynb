```python
# delete duplicates and save as the same dataframe
df = df.drop_duplicates(subset=['article_text']) # 
```

### sort

```python
df = df.sort_values('source', ascending=True)
df
```

### count

```python
# count how many sources are in the data

df['source'].value_counts().reset_index()
```

```python
# count # of articles by sources

df_count_source = df.groupby('source')['article_text'].count().reset_index()
df_count_source
```

```python
# add multiple groups to count

df_count_source_date = df.groupby(['source', 'date_published'])['article_text'].count().reset_index()
df_count_source_date
```

```python
# count by publish dates

df.groupby('date_published')['article_text'].count().reset_index()
```

### random sample by groups

```python
df_source_rndm1 = df.groupby('source').apply(lambda x: x.sample(1, replace=False)).reset_index(drop=True)
df_source_rndm1
```
